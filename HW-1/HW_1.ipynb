```python
# Part C
import numpy as np
import matplotlib.pyplot as plt
import scipy.io

def plot_frequency_spectrum(signal, fs):
    # Fourier
    n = len(signal)
    fft_result = np.fft.fft(signal)
    fft_magnitude = np.abs(fft_result)[:n//2]

    frequencies = np.fft.fftfreq(n, 1/fs)[:n//2]

    plt.figure(figsize=(10, 4))
    plt.plot(frequencies, fft_magnitude)
    plt.title('Frequency Spectrum')
    plt.xlabel('Frequency [Hz]')
    plt.ylabel('Magnitude')
    plt.grid(True)
    plt.show()

    return frequencies, fft_magnitude

!gdown 1BAhLkLr74w0sGdHZrCecwlQ79NlSNx5A
data = scipy.io.loadmat('109.mat')

fs = 48000
signal = data['X109_DE_time']


frequencies, fft_magnitude = plot_frequency_spectrum(signal, fs)
dominant_frequency = frequencies[np.argmax(fft_magnitude)]
print(f"Dominant Frequency: {dominant_frequency} Hz")
```

# Iris Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# Part 1
def describe_iris_dataset():
    iris = load_iris()
    print("Iris Dataset Description:")
    print(iris.DESCR)
    print("Feature:")
    print(iris.feature_names)
    print("Class:")
    print(iris.target_names)

# Part 2
def load_iris_data():
    iris = load_iris()
    df = pd.DataFrame(iris.data, columns=iris.feature_names)
    df['species'] = iris.target
    return df

# Part 3
def split_data(df):
    X = df.iloc[:, :-1]
    y = df.iloc[:, -1]
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    return X_train, X_test, y_train, y_test

describe_iris_dataset()
df = load_iris_data()
print("\n Five first row:")
print(df.head())

X_train, X_test, y_train, y_test = split_data(df)
print("\n Number Of traning sample:", len(X_train))
print("Number of test:", len(X_test))
```

```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# Part 1
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target

# Part 2
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, :-1], df.iloc[:, -1], test_size=0.3, random_state=42)

# Part 3
train_df = pd.DataFrame(X_train, columns=iris.feature_names)
train_df['species'] = y_train
train_df['dataset_type'] = 'train'

test_df = pd.DataFrame(X_test, columns=iris.feature_names)
test_df['species'] = y_test
test_df['dataset_type'] = 'test'

# Part 4
combined_df = pd.concat([train_df, test_df], axis=0)

print("DataFrame ادغام شده:")
print(combined_df)

combined_df.to_csv('combined_iris_dataset.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# Part 1
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target

# Part 2
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, :-1], df.iloc[:, -1], test_size=0.3, random_state=42)

# Part 3
train_df = pd.DataFrame(X_train, columns=iris.feature_names)
train_df['species'] = y_train
train_df['dataset_type'] = 'train'

test_df = pd.DataFrame(X_test, columns=iris.feature_names)
test_df['species'] = y_test
test_df['dataset_type'] = 'test'

# Part 4
combined_df = pd.concat([train_df, test_df], axis=0)

# Show Distribution in 2D
plt.figure(figsize=(10, 6))
sns.scatterplot(data=combined_df, x='sepal length (cm)', y='sepal width (cm)', hue='species', style='dataset_type')
plt.title('2D Distribution')
plt.show()

# Show Distribution in 3D
fig = px.scatter_3d(combined_df, x='sepal length (cm)', y='sepal width (cm)', z='petal length (cm)', color='species', symbol='dataset_type')
fig.update_layout(title='3D Distrbution')
fig.show()

# Heat Map
plt.figure(figsize=(10, 6))
sns.heatmap(combined_df.iloc[:, :-2].corr(), annot=True, cmap='coolwarm')
plt.title('Heat Map')
plt.show()

# Probiblity Density
features = iris.feature_names
plt.figure(figsize=(15, 10))
for i, feature in enumerate(features):
    plt.subplot(2, 2, i+1)
    sns.kdeplot(data=train_df, x=feature, label='Train', fill=True)
    sns.kdeplot(data=test_df, x=feature, label='Test', fill=True)
    plt.title(f'Probility Density {feature}')
    plt.legend()
plt.tight_layout()
plt.show()
```
