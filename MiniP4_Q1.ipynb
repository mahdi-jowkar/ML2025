# MiniP4-Q1



## import library



```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

## Nueron Definition

```python
#define muculloch pitts
class McCulloch_Pitts_neuron():

  def __init__(self , weights , threshold):
    self.weights = weights    #define weights
    self.threshold = threshold    #define threshold

  def model(self , x):
    #define model with threshold
    if self.weights @ x >= self.threshold:
        return 1
    else:
        return 0
```

## neurons

```python
neur1 = McCulloch_Pitts_neuron([2, -1], +2)
neur2 = McCulloch_Pitts_neuron([-2, -1], -6)
neur3 = McCulloch_Pitts_neuron([0, 1], +0)
neur4 = McCulloch_Pitts_neuron([1,1,1], +3)
```

# Random Number and Space Creation

```python
randomnum = np.random.RandomState(56) #Studend Number 40316956
xs = randomnum.uniform(0,4,2000)
ys = randomnum.uniform(-1,3,2000)
classes = ['r'] * 2000  # Creating 2000 default points to be changed later
```

# Simple McCulloch_Pitts Classification

```python
def classify(Threshold):


        for i in range(0,2000) :

                input = np.asmatrix([[ xs[i] ] , [ ys[i] ]])

                #hidden Layer 1
                z1 = neur1.model(input)
                z2 = neur2.model(input)
                z3 = neur3.model(input)

                #Output Layer
                z4 = neur4.model(np.array([z1, z2, z3]))
                # Inside Triangle of 1 and outside otherwise

                if z4 >= Threshold :
                        classes[i]='g'
                else:
                        classes[i]='r'

classify(Threshold=1)

#PLOT
plt.scatter( xs, ys  , c=classes)
plt.title('Classification using mcculloch pitts neuron')
plt.xlabel('xs')
plt.ylabel('ys')
plt.grid(True)
# The boundaries are defined by the first three neurons:
# neur1: 2*x - y >= 2  => y <= 2*x - 2
# neur2: -2*x - y >= -6 => y <= -2*x + 6
# neur3: y >= 0
x_vals = np.linspace(0, 4, 100)
# Line 1: y = 2*x - 2
plt.plot(x_vals, 2 * x_vals - 2, color='blue', linestyle='-', label='Boundary 1')
# Line 2: y = -2*x + 6
plt.plot(x_vals, -2 * x_vals + 6, color='green', linestyle='-', label='Boundary 2')
# Line 3: y = 0
plt.axhline(y=0, color='purple', linestyle='-', label='Boundary 3')
plt.xlim(0, 4)
plt.ylim(-1, 3)
plt.legend()
plt.show()
```

# Investigating Use Of Relu

## Relu MLP definition

```python
class ReluMLP():

  def __init__(self , weights , threshold):
    self.weights = weights  #define weights
    self.threshold = threshold    #define threshold

  def model(self , x):
    #define model with threshold
    func=self.weights @ x+self.threshold
    return max(0 , func.item() )
```

## Using Relu

```python
neur1 = ReluMLP([+2, -1], -2)
neur2 = ReluMLP([-2, -1], 6)
neur3 = ReluMLP([+0, +1], 0)
neur4 = ReluMLP([+1,1,1], -3)

classify(Threshold=1)

plt.scatter( xs , ys  , c=classes)
plt.title('Relu Activation but Weights kept the same')
plt.xlabel('xs')
plt.ylabel('ys')
plt.grid(True)
# The boundaries are defined by the first three neurons:
# neur1: 2*x - y >= 2  => y <= 2*x - 2
# neur2: -2*x - y >= -6 => y <= -2*x + 6
# neur3: y >= 0
x_vals = np.linspace(0, 4, 100)
# Line 1: y = 2*x - 2
plt.plot(x_vals, 2 * x_vals - 2, color='blue', linestyle='-', label='Boundary 1')
# Line 2: y = -2*x + 6
plt.plot(x_vals, -2 * x_vals + 6, color='green', linestyle='-', label='Boundary 2')
# Line 3: y = 0
plt.axhline(y=0, color='purple', linestyle='-', label='Boundary 3')
plt.xlim(0, 4)
plt.ylim(-1, 3)
plt.legend()
plt.show()
```

# Investigating Use of Sigmoid

## Sigmoid MLP definition

```python
class SigmoidMLP():

  def __init__(self , weights , threshold):
    self.weights = weights    #define weights
    self.threshold = threshold    #define threshold

  def model(self , x):
    #define model with threshold
    func=self.weights @ x+self.threshold
    return 1/(1+np.exp(-func.item()))
```

## Using Sigmoid

```python
neur1 = SigmoidMLP([+2, -1], -2)
neur2 = SigmoidMLP([-2, -1], 6)
neur3 = SigmoidMLP([+0, +1], 0)
neur4 = SigmoidMLP([+1,1,1], -3)

classify(Threshold=1)

plt.scatter( xs , ys  , c=classes)
plt.title('Sigmoid Activation but Weights kept the same')
plt.xlabel('xs')
plt.ylabel('ys')
plt.grid(True)
# The boundaries are defined by the first three neurons:
# neur1: 2*x - y >= 2  => y <= 2*x - 2
# neur2: -2*x - y >= -6 => y <= -2*x + 6
# neur3: y >= 0
x_vals = np.linspace(0, 4, 100)
# Line 1: y = 2*x - 2
plt.plot(x_vals, 2 * x_vals - 2, color='blue', linestyle='-', label='Boundary 1')
# Line 2: y = -2*x + 6
plt.plot(x_vals, -2 * x_vals + 6, color='green', linestyle='-', label='Boundary 2')
# Line 3: y = 0
plt.axhline(y=0, color='purple', linestyle='-', label='Boundary 3')
plt.xlim(0, 4)
plt.ylim(-1, 3)
plt.legend()
plt.show()
```
